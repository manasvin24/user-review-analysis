```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('../data/BDFoodSent-334k.csv')
```

```python
df.columns
```

```python
def describe_columns(dataframe):
    """
    Provides detailed information about each column in the dataframe
    """
    
    for col in dataframe.columns:
        print(f"\n{'='*80}")
        print(f"Column: {col}")
        print(f"{'='*80}")
        print(f"Data Type: {dataframe[col].dtype}")
        print(f"Total Count: {dataframe[col].count()}")
        print(f"Missing Values: {dataframe[col].isna().sum()}")
        print(f"Unique Values: {dataframe[col].nunique()}")
        print(f"Sample Values: {dataframe[col].dropna().unique()[:5]}")
        # Basic statistics for numeric columns
        if dataframe[col].dtype in ['int64', 'float64']:
            print(f"\nNumeric Statistics:")
            print(f"  Min: {dataframe[col].min()}")
            print(f"  Max: {dataframe[col].max()}")
            print(f"  Mean: {dataframe[col].mean():.2f}")
            print(f"  Median: {dataframe[col].median():.2f}")
            print(f"  Std Dev: {dataframe[col][0]}")
        
        print()
    
    print("="*80)

# Call the function
describe_columns(df)
```
